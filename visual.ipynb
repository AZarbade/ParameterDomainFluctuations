```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('data/SIDI_Full.csv')
data['idx'] = data.index
data.head(3)
```

```python
fig = px.scatter_matrix(data, color='LD', color_continuous_scale=px.colors.qualitative.D3)
fig.show()
```

```python
fig_3d = px.scatter_3d(
    data, x='a (degrees)', y='$ (degrees)', z='DI',
    color='LD',
    color_continuous_scale=px.colors.qualitative.D3,
    size='Velocity (km/s)'
)

fig_3d.show()
```

```python
from umap import UMAP

umap_2d = UMAP(2)
proj_2d = umap_2d.fit_transform(data)

fig_2d = px.scatter(
    proj_2d, x=0, y=1,
    color=data['LD'],
    color_continuous_scale=px.colors.qualitative.D3
)

fig_2d.show()
```

```python
umap_3d = UMAP(3)
proj_3d = umap_3d.fit_transform(data)

fig_3d = px.scatter_3d(
    proj_3d, x=0, y=1, z=data['DI'],
    color=data['LD'],
    color_continuous_scale=px.colors.qualitative.D3
)

fig_3d.show()
```

```python
LD_1 = data.drop('LD', axis=1)
LD_1.head()
```

```python
fig = px.scatter_3d(
    z=LD_1['DI'],
    x=LD_1['a (degrees)'],
    y=LD_1['$ (degrees)'],
    color=LD_1['Velocity (km/s)']
)
fig.show()
```

```python
umap_3d = UMAP(3)
proj_3d = umap_3d.fit_transform(LD_1)

fig_3d = px.scatter_3d(
    proj_3d, x=0, y=1, z=LD_1['DI'],
    color=LD_1['Velocity (km/s)'],
    color_continuous_scale=px.colors.qualitative.D3
)

fig_3d.show()
```

```python
umap_2d = UMAP(2)
proj_2d = umap_2d.fit_transform(LD_1)

fig_2d = px.scatter(
    proj_2d, x=0, y=1,
    color=data['Velocity (km/s)'],
    color_continuous_scale=px.colors.qualitative.D3
)

fig_2d.show()
```
